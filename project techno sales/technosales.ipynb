```python
profit_by_category_brand = data.groupby(['Category', 'Brand'])['Profit'].sum().reset_index()
profit_by_category_brand
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
profit_by_category_brand.plot(kind='bar', ax=ax, color=['red', 'orange', 'red'])
ax.set_title('Total Profit Distributed by Category and Brand', fontsize=16)
ax.set_xlabel('Category', fontsize=12)
ax.set_ylabel('Total Profit', fontsize=12)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.legend(title='Brand')
plt.tight_layout()
plt.show()
```

```python
profit_by_category_brand = data.groupby(['Category', 'Brand'])['Profit'].sum().unstack().fillna(0)
profit_by_category_brand
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
profit_by_category_brand.plot(kind='bar', ax=ax1, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
ax1.set_title('Total Profit by Category and Brand', fontsize=14)
ax1.set_xlabel('Category', fontsize=12)
ax1.set_ylabel('Total Profit', fontsize=12)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
ax1.legend(title='Brand')
total_profit_by_category = data.groupby('Category')['Profit'].sum()
total_profit_by_category.plot(kind='pie', ax=ax2, autopct='%1.1f%%', colors=['#1f77b4', '#ff7f0e', '#2ca02c'])
ax2.set_title('Total Profit by Category', fontsize=14)
ax2.set_ylabel('')
plt.tight_layout()
plt.show() 
```
